```python
def process(args):
	text = args[1]
	print(args[0]+"is processed")
	for p in punctuation:
		text = text.replace(p, " ")

	doc = Doc(text)
	doc.segment(segmenter)

	doc.tag_morph(morph_tagger)

	for token in doc.tokens:
		token.lemmatize(morph_vocab)

	return [args[0], [token.lemma for token in doc.tokens if token.lemma not in stop_words]]
```

```python
import numpy as np
from numba import jit
from math import log, e
from scipy.stats import entropy
n = len(answers)




@jit(nopython=True)
def character_answer(spec_vector, question):
    if spec_vector[question]:
      return 1/(1+np.exp(-7*(spec_vector[question]-0.58)))
    else:
      return 0.0

@jit(nopython=True, parallel=True)
def calculate_spec_probability(spec, questions_so_far, answers_so_far, answer_id):
    
    P_spec = 1 / answer_id.shape[0]
    spec_vec = answer_id[spec]
    
    P_answers_given_spec = 1
    P_answers_given_not_spec = 1
    for question, answer in zip(questions_so_far, answers_so_far):
        P_answers_given_spec *= max(
            1 - abs(answer - character_answer(spec_vec, question)), 0.001)
        P_answer_not_spec = np.mean(np.array([1 - abs(answer - character_answer(answer_id[i], question))
                                          for i in range(n)
                                          if i != spec], dtype=np.float64))
        P_answers_given_not_spec *= max(P_answer_not_spec, 0.001)

    
    P_answers = P_spec * P_answers_given_spec + \
        (1 - P_spec) * P_answers_given_not_spec

   
    P_spec_given_answers = (
        P_answers_given_spec * P_spec) / P_answers

    return P_spec_given_answers
  
@jit(nopython=True, parallel=True)
def suggest_question(questions_so_far, answers_so_far, answer_id, choices):

  max_prob = 0.0
  question = 0
  for i in choices:
    print(i)
    if np.any(questions_so_far == i):
      continue
    m = 0.0
    for ans in np.array([0, 0.25, 0.5, 0.75, 1]):
      new_questions = np.append(questions_so_far, np.array([i]))
      new_answers = np.append(answers_so_far, np.array(ans))
      vals = np.array([
          calculate_spec_probability(j, new_questions, new_answers, answer_id) for j in range(answer_id.shape[0])
      ])
      m += np.max(vals)/4.0
    if max_prob < m:
      max_prob = m
      question = i 
  return question

```
